```python
df = pd.read_csv('/content/vgames2.csv')

print (df['Genre'].value_counts())
print (df['Year'].value_counts())
print (df['Platform'].value_counts())
print (df[df.duplicated(['Name','Platform'])])
print (df.isnull().sum())
```

```python
df = df.fillna(0)
df['Genre'].replace(0, 'Other', inplace = True)
df = df.astype({'Year': 'int'})
before_mil = ((df['Year'] < 1000) & (df['Year'] > 22))
after_mil = df['Year'] < 22
df.Year[before_mil] = df.Year[before_mil] + 1900
df.Year[after_mil] = df.Year[after_mil] + 2000
```

```python
#df['NA_Sales'] = df['NA_Sales'].replace({'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(float)
#df['EU_Sales'] = df['EU_Sales'].replace({'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(float)
#df['JP_Sales'] = df['JP_Sales'].replace({'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(float)
#df['Other_Sales'] = df['Other_Sales'].replace({'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(float)
#df
```

```python
df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']] = df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].apply(lambda x: x.str.replace('K',''))
df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']] = df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].apply(lambda x: x.str.replace('M',''))
df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']] = df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].astype(float)
df
```

```python
df['Sales_Total'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']
df = df.sort_values(by=['Sales_Total'], axis=0, ascending=False)
df3 = df.head(50)
df2 = df[df['Sales_Total'] < 75]
df2
```

```python
df_yr = df.groupby(['Year']).sum()
df_gr = df.groupby(['Genre']).sum()

print(df_yr)
print(df_gr)
```
